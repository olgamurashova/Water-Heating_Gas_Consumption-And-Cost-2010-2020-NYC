```python
#Checking Null values in the dataset
df.isnull().sum()
```

```python
#Renaming the column name
df.rename(columns = {'Consumption (Therms)': "Consumption_Therms"})
```

```python
#Filling the Null values by central tendency (Mean, Mode, Median)
#Mode is used for categorical data
cols = ['Development Name', 'Location', 'AMP #', 'Service Start Date', 'Service End Date']
df[cols] = df[cols].fillna(df.mode().iloc[0])
```

```python
#mean and median are used for numerical values
cols = ['TDS #','Estimated', '# days']
df[cols] = df[cols].fillna(df.median().iloc[0])
```

```python
df.isnull().sum()
```

```python
#Grouping the data
#Grouping the data by development, borough and unique identifier and calculating the total charges for each development
df.groupby(['Development Name', 'Borough', 'TDS #'])['Current Charges'].sum()
```

```python
#Creating bins to divide consumption of gas range into 5 categories 
pd.cut(df['Consumption (Therms)'], bins = 5).value_counts()
```
